**Data Cleaning and Wrangling**

```python
#merge with customers
df=df_sales.merge(
    df_customers,
    how='left',
    left_on='Customer Name Index',
    right_on='Customer Index'
)
```

```python
df.head()
```

```python
#merge with Products
df=df.merge(
    df_products,
    how='left',
    left_on='Product Description Index',
    right_on='Index'
)
```

```python
df.head()
```

```python
#merge with regions 
df= df.merge(
    df_regions,
    how='left',
    left_on='Delivery Region Index',
    right_on='id'
)
```

```python
df.head()
```

```python
#merge with state regions 
df= df.merge(
    df_state_reg[["State Code","Region"]],
    how='left',
    left_on='state_code',
    right_on='State Code'
)
```

```python
# merge with budgets
df=df.merge(
    df_budgets,
    how='left',
    on='Product Name'
)
```

```python
df.head()
```

```python
#Customer Index
#Index
# ID
#State Code
# Clean up redundant columns 
cols_to_drop=['Customer Index','Index','id','State Code']
df=df.drop(columns=cols_to_drop,errors='ignore')
df.head()
```

```python
#covert all columns to lower case for easy access and for better consistency
df.columns=df.columns.str.lower()
df.columns.values
```

```python
#keep the important columns and delete the coulmns that we dont need 
cols_to_keep=[
    'ordernumber',
    'orderdate',
    'customer names',
    'channel',
    'product name',
    'order quantity',
    'unit price',
    'line total',
    'total unit cost',
    'state_code',
    'county',
    'state',
    'region',
    'latitude',
    'longitude',
    '2017 budgets'
]
```

```python
df=df[cols_to_keep]
```

```python
df.head()
```

```python
#Rename the columns
df=df.rename(columns={
    'ordernumber':'order_number',
    'orderdate':'order_date',
    'customer name':'customer_name',
    'product name': 'product_name',
    'order quantity':'order_quantity',
    'unit price':'unit_price',
    'line total':'revenue',
    'total_unit_cost': 'cost',
    'state_code': 'state',
    'state': 'state_name',
    'latitude':'lat',
    'longitude':'lon',
    '2017 budgets':'budget'
})
df.head(1)
```

```python
#Blank out budgets for non-2017 orders 
df.loc[df['order_date'].dt.year!=2017,'budget'] = pd.NA

#line total is revenue
df[['order_date','product_name','revenue','budget']].head(5)
```

**Feature engineering**

```python
df['total_cost']=df['order_quantity']*df['total unit cost']
```

```python
df['profit']=df['revenue']-df['total_cost']

df['profit_margin_pct'] = df['profit']/df['revenue']*100
```

```python
import matplotlib.pyplot as plt
df['order_month']=df['order_date'].dt.to_period('M')

monthly_sales=df.groupby('order_month')['revenue'].sum()

plt.figure(figsize=(15,4))

monthly_sales.plot(marker='o',color='navy')

from matplotlib.ticker import FuncFormatter
formatter=FuncFormatter(lambda x,pos:f'{x/1e6:.1f}M')
plt.gca().yaxis.set_major_formatter(formatter)
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Revenue (Millions)')

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
#dataframe is df and converting datetime 
df['order_date']=pd.to_datetime(df['order_date'])
#Remove records from january and february 2018
df_new=df[~((df['order_date'].dt.year==2018) & (df['order_date'].dt.month.isin([1,2])))]
```

```python
#coverting the order_date to datetime
df_new['order_date']=pd.to_datetime(df_new['order_date'])

#extract month name for grouping  
df_new['month']=df_new['order_date'].dt.strftime('%B')

#aggregate revenue by month across all years
monthly_trend=df_new.groupby('month')['revenue'].sum().reset_index()

#sort months in calender order 
month_order=['January','February','March','April','May','June','July','August','September','October','November','December']
monthly_trend['month']=pd.Categorical(monthly_trend['month'],categories=month_order,ordered=True)
monthly_trend=monthly_trend.sort_values('month')

#create a line chart
plt.figure(figsize=(10,6))
sns.lineplot(data=monthly_trend,x='month',y='revenue',marker='o')
plt.title('Overall Monthly Sales trend (All Years)',fontsize=14)
plt.xlabel('Month',fontsize=12)
plt.ylabel('Revenue($)',fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

#show the plot 
plt.show()
```

```python
# Step 1: Aggregate revenue by product
top_products = df.groupby('product_name')['revenue'].sum().nlargest(10).reset_index()
top_products = top_products.sort_values('revenue', ascending=False)

# Step 2: Create the bar chart
plt.figure(figsize=(12, 6))
barplot = sns.barplot(data=top_products, x='product_name', y='revenue', palette='viridis')

# Step 3: Add labels on top of bars
for index, row in top_products.iterrows():
    barplot.text(index, row['revenue'], f"${row['revenue']:,.0f}", color='black', ha='center', va='bottom', fontsize=10)

# Step 4: Customize chart
plt.title('Top 10 Products by Total Revenue', fontsize=14)
plt.xlabel('Product Name', fontsize=12)
plt.ylabel('Revenue ($)', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
# Step 1: Get bottom 10 products by total revenue
bottom_products = df.groupby('product_name')['revenue'].sum().nsmallest(10).reset_index()

# Step 2: Sort for clear visualization
bottom_products = bottom_products.sort_values('revenue', ascending=True)

# Step 3: Create the bar chart
plt.figure(figsize=(12, 6))
barplot = sns.barplot(data=bottom_products, x='product_name', y='revenue', palette='rocket')

# Step 4: Add value labels above each bar
for index, row in bottom_products.iterrows():
    barplot.text(index, row['revenue'], f"${row['revenue']:,.0f}", color='black', ha='center', va='bottom', fontsize=10)

# Step 5: Customize appearance
plt.title('Bottom 10 Products by Total Revenue', fontsize=14)
plt.xlabel('Product Name', fontsize=12)
plt.ylabel('Revenue ($)', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
#sales nby channel 
chan_sales=df.groupby('channel')['revenue'].sum().sort_values(ascending=True)
plt.figure(figsize=(5,5))
plt.pie(
    chan_sales.values,
    labels=chan_sales.index,
    autopct='%1.1f%%',
    colors=sns.color_palette('coolwarm')
)
plt.title('Total Sales By channel')
plt.tight_layout()
plt.show()
```

```python
#average order value Distribution
aov=df.groupby('order_number')['revenue'].sum()
plt.figure(figsize=(12,4))
plt.hist(
    aov,
    bins=50,
    color='skyblue',
    edgecolor='black'
)
plt.title('Distribution of Average Order Value')
plt.xlabel('Order Value (USD)')
plt.ylabel('Number of Orders')
plt.tight_layout()
plt.show()
```

```python
# Set up the plot canvas
plt.figure(figsize=(12, 6))

# Create boxplot for unit price distribution by product
sns.boxplot(data=df, x='product_name', y='unit_price', palette='Spectral')

# Add chart aesthetics
plt.title('Unit Price Distribution by Product', fontsize=14)
plt.xlabel('Product Name', fontsize=12)
plt.ylabel('Unit Price ($)', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
# Step 1: Aggregate revenue by state
top_states = df.groupby('state_name')['revenue'].sum().nlargest(10).reset_index()

# Step 2: Sort for readability
top_states = top_states.sort_values('revenue', ascending=False)

# Step 3: Create the bar chart
plt.figure(figsize=(12, 6))
barplot = sns.barplot(data=top_states, x='state_name', y='revenue', palette='crest')

# Step 4: Add labels above bars
for index, row in top_states.iterrows():
    barplot.text(index, row['revenue'], f"${row['revenue']:,.0f}", color='black', ha='center', va='bottom', fontsize=10)

# Step 5: Final touches
plt.title('Top 10 States by Total Revenue', fontsize=14)
plt.xlabel('State', fontsize=12)
plt.ylabel('Revenue ($)', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
# Step 1: Aggregate total revenue by customer
customer_revenue = df.groupby('customer names')['revenue'].sum().reset_index()

# Step 2: Get top and bottom 10 customers
top_10 = customer_revenue.nlargest(10, 'revenue')
bottom_10 = customer_revenue.nsmallest(10, 'revenue')

# Step 3: Combine and sort
combined = pd.concat([top_10, bottom_10]).sort_values('revenue', ascending=True).reset_index(drop=True)

# Step 4: Create horizontal bar chart
plt.figure(figsize=(10, 8))
barplot = sns.barplot(data=combined, y='customer names', x='revenue', palette='coolwarm')

# Step 5: Add value labels beside bars
for index, row in combined.iterrows():
    barplot.text(row['revenue'], index, f"${row['revenue']:,.0f}", va='center', ha='left', fontsize=10)

# Final touches
plt.title('Top & Bottom 10 Customers by Total Revenue', fontsize=16)
plt.xlabel('Revenue ($)', fontsize=12)
plt.ylabel('Customer Name', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Step 1: Aggregate total revenue and average profit margin per customer
customer_segmentation = df.groupby('customer names').agg({
    'revenue': 'sum',
    'profit_margin_pct': 'mean'
}).reset_index()

# Step 2: Create scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=customer_segmentation,
                x='revenue',
                y='profit_margin_pct',
                hue='profit_margin_pct',
                size='revenue',
                sizes=(50, 300),
                palette='viridis',
                alpha=0.8,
                legend=False)

# Step 3: Highlight top 5 customers by revenue
top_customers = customer_segmentation.nlargest(5, 'revenue')
for _, row in top_customers.iterrows():
    plt.text(row['revenue'], row['profit_margin_pct'], row['customer names'],
             fontsize=9, weight='bold', color='darkblue')

# Final chart touch-ups
plt.title('Customer Segmentation: Revenue vs Profit Margin', fontsize=14)
plt.xlabel('Total Revenue ($)', fontsize=12)
plt.ylabel('Average Profit Margin (%)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Select relevant numerical columns for correlation
numeric_cols = ['order_quantity', 'unit_price', 'revenue', 'total unit cost', 
                'budget', 'total_cost', 'profit', 'profit_margin_pct']

# Drop rows with NaN values in selected columns
df_corr = df[numeric_cols].dropna()

# Compute correlation matrix
corr_matrix = df_corr.corr()

# Set up the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap="coolwarm", center=0, linewidths=0.5)

# Add titles and layout improvements
plt.title("Correlation Heatmap of Sales Metrics", fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
